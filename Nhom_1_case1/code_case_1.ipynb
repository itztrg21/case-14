```python
import numpy as np
```

##### 1.ĐỌC FILE CSV ( SỬ DỤNG GENFROMTEXT )

```python
dtype = [('date', 'U10'), ('price', 'f8')]
hcm = np.genfromtxt("C:/Users/admin/Downloads/gold_price_hcm.csv", delimiter=",", dtype=dtype, skip_header=1)
hn  = np.genfromtxt("C:/Users/admin/Downloads/gold_price_hanoi.csv", delimiter=",", dtype=dtype, skip_header=1)
print("HCM sample:", hcm[:2])
print("HN  sample:", hn[:2])
print()
```

 ##### 2.TÁCH CỘT NGÀY VÀ GIÁ RIÊNG

```python
hcm_dates = hcm['date']
hcm_prices = hcm['price']
hn_dates = hn['date']
hn_prices = hn['price']
#vd
print("date_hcm:", hcm_dates[:3])
print("price_hcm:", hcm_prices[:3])
print()
```

##### 3.CHUYỂN CỘT NGÀY SANG KIỂU DỮ LIỆU DATETIME64

```python
hcm_dates=hcm_dates.astype('datetime64[D]')
hn_dates=hn_dates.astype('datetime64[D]')
```

##### 4.DTYPE, SHAPE VÀ HEAD ( 5 GIÁ TRỊ ĐẦU TIÊN )

###### Giá vàng Hồ Chí Minh

```python
print("HCM dtype:", hcm.dtype, "shape:", hcm.shape)
print("5 Giá trị đầu tiên ở Hồ Chí Minh: ", hcm_prices[:5])
print("5 Giá trị cuối ở Hồ Chí Minh: ", hcm_prices[-5:])
```

###### Giá vàng Hà Nội

```python
print("HCM dtype:", hn.dtype, "shape:", hn.shape)
print("5 Giá trị đầu tiên ở Hà Nội: ", hn_prices[:5])
print("5 Giá trị cuối ở Hà Nội: ", hn_prices[-5:])
```

##### 5.SLICING: LẤY 7 GIÁ TRỊ ĐẦU, GIỮA, CUỐI

###### Giá vàng Hồ Chí Minh

```python

hcm_first7 = hcm_prices[:7]
hcm_mid7   = hcm_prices[len(hcm_prices)//2 - 3 : len(hcm_prices)//2 + 4]
hcm_last7  = hcm_prices[-7:]
print('7 Giá trị đầu của đữ liệu giá vàng Hồ Chí Minh: ',hcm_first7)
print('7 Giá trị giữa của dữ liệu giá vàng Hồ Chí Minh: ',hcm_mid7)
print('7 Giá trị cuối của dữ liệu giá vàng Hồ Chí Minh: ',hcm_last7)
```

###### Giá vàng Hà Nội

```python

hn_first7 = hn_prices[:7]
hn_mid7   = hn_prices[len(hcm_prices)//2 - 3 : len(hcm_prices)//2 + 4]
hn_last7  = hn_prices[-7:]
print('7 Giá trị đầu của đữ liệu giá vàng Hà Nội: ',hn_first7)
print('7 Giá trị giữa của dữ liệu giá vàng Hà Nội: ',hn_mid7)
print('7 Giá trị cuối của dữ liệu giá vàng Hà Nội: ',hn_last7)
```

##### 6.BOOLEAN: TÌM NGÀY GIÁ VÀNG HCM CAO HƠN HN

```python

gia_tri_can_tim = hcm_prices > hn_prices
higher_dates = hcm_dates[gia_tri_can_tim]
hcm_high     = hcm_prices[gia_tri_can_tim]
hn_low       = hn_prices[gia_tri_can_tim]
for i in range(len(higher_dates)):
    d = higher_dates[i]
    h = hcm_high[i]
    n = hn_low[i]
    print(f"Ngày {d}: HCM = {h}, HN = {n}")

```

##### 7.TOP 5 NGÀY CÓ GIÁ CAO NHẤT Ở 2 KHU VỰC HÀ NỘI VÀ HỒ CHÍ MINH

###### Hồ Chí Minh

```python
id_top5_hcm = np.argsort(hcm['price'])[-5:]
print("TOP 5 HCM:")
for i in id_top5_hcm:
    print(hcm['date'][i],hcm['price'][i])
```

###### Hà Nội

```python
id_top5_hn = np.argsort(hn['price'])[-5:]
print("TOP 5 HN:")
for i in id_top5_hn:
    print(hn['date'][i],hn['price'][i])
```

##### 8.CHÊNH LỆCH GIÁ VÀNG TỪNG NGÀY ( LẤY TẠM DỮ LIỆU 10 GIÁ TRỊ ĐẦU TIÊN )

```python
diff = hcm_prices - hn_prices
diff_table = np.column_stack([
    hcm_dates.astype(str), 
    hcm_prices,
    hn_prices,
    diff
])
print('  Ngày         Hồ Chí Minh    Hà Nội     Chênh lệch giá    ')
for row in diff_table[:10]:
    print(row)
```

##### 9.SỬ DỤNG NP.DIFF ĐỂ QUAN SÁT CHÊNH LỆCH GIÁ GIỮA CÁC NGÀY

###### Hồ Chí Minh ( 10 ngày đầu tiên trong dataset )

```python
gia_cu = hcm_prices[:-1]
gia_moi = hcm_prices[1:]
change = gia_cu - gia_moi
new_change = hcm_dates[1:]
table = np.column_stack([
    new_change.astype(str),
    gia_cu,
    gia_moi,
    change
])
for row in table[:10]:
    print(row)
```

###### Hà Nội ( 10 ngày đầu tiên trong dataset )

```python
gia_cu = hn_prices[:-1]
gia_moi = hn_prices[1:]
change = gia_cu - gia_moi
new_change = hn_dates[1:]
table = np.column_stack([
    new_change.astype(str),
    gia_cu,
    gia_moi,
    change
])
for row in table[:10]:
    print(row)
```

##### 10.TÍNH STD(độ lệch chuẩn), MAX MIN, ARGMAX ARGMIN

###### Hồ Chí Minh

```python
print("HCM mean:", np.mean(hcm_prices))
print("HCM std:", np.std(hcm_prices))
print("HCM max:", np.max(hcm_prices), "Ngày", hcm_dates[np.argmax(hcm_prices)])
print("HCM min:", np.min(hcm_prices), "Ngày", hcm_dates[np.argmin(hcm_prices)])
```

###### Hà Nội 

```python

print("HN  mean:", np.mean(hn_prices))
print("HN std:", np.std(hn_prices))
print("HN max:", np.max(hn_prices), "Ngày", hn_dates[np.argmax(hn_prices)])
print("HN min:", np.min(hn_prices), "Ngày", hn_dates[np.argmin(hn_prices)])
```

##### 12.VẼ BIỂU ĐỒ LINE HOẶC BAR MINH HỌA


```python
import numpy as np
import matplotlib.pyplot as plt
```

###### Chart line:So sánh giá vàng HCM và Hà Nội theo thời gian

```python
plt.figure(figsize=(12, 6))
plt.plot(hcm_dates, hcm_prices, label="HCM", marker="o")
plt.plot(hn_dates, hn_prices, label="Hà Nội", marker="o")
plt.title("So sánh giá vàng HCM và Hà Nội")
plt.xlabel("Ngày")
plt.ylabel("Giá vàng")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

###### Chart bar:So sánh giá vàng từng ngày giữa 2 thành phố.

```python
N = 10
dates_sample = hcm_dates[:N]
hcm_sample   = hcm_prices[:N]
hn_sample    = hn_prices[:N]
x = np.arange(N)
width = 0.35
plt.figure(figsize=(12, 6))
plt.bar(x - width/2, hcm_sample, width=width, label="HCM")
plt.bar(x + width/2, hn_sample,  width=width, label="Hà Nội")
plt.title("Biểu đồ Bar so sánh giá vàng (10 ngày đầu)")
plt.xlabel("Ngày")
plt.ylabel("Giá vàng")
plt.xticks(x, dates_sample, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
